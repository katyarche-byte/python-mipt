```python
import pygame

pygame.init()
screen = pygame.display.set_mode((960, 768))
pygame.display.set_caption("Панда")

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill((250, 210, 150))

    BLACK = (0, 0, 0)
    WHITE = (255, 255, 255)
    GREEN = (4, 111, 65)
    
    #передаёт, где находится мышь
    mouse_x, mouse_y = pygame.mouse.get_pos()
    
    # Стебель (справа)
    pygame.draw.rect(screen, GREEN, (480, 430, 40, 120))
    pygame.draw.rect(screen, GREEN, (480, 300, 40, 120))
    pygame.draw.rect(screen, GREEN, (480, 170, 40, 120))
    
    # Стебель (слева)
    pygame.draw.rect(screen, GREEN, (350, 410, 30, 100))
    pygame.draw.rect(screen, GREEN, (350, 300, 30, 100))

    # Листья справа
    pygame.draw.line(screen, GREEN, (520, 190), (740, 120), 4) # Верхняя ветвь
    pygame.draw.line(screen, GREEN, (740, 120), (755, 100), 3)
    pygame.draw.line(screen, GREEN, (745, 118), (760, 98), 3)
    pygame.draw.line(screen, GREEN, (710, 125), (725, 105), 3)
    pygame.draw.line(screen, GREEN, (715, 123), (730, 103), 3)
    pygame.draw.line(screen, GREEN, (680, 132), (695, 112), 3)
    pygame.draw.line(screen, GREEN, (685, 130), (700, 110), 3)

    pygame.draw.line(screen, GREEN, (520, 320), (320, 220), 4) # Средняя ветвь
    pygame.draw.line(screen, GREEN, (320, 220), (310, 200), 3)
    pygame.draw.line(screen, GREEN, (325, 218), (315, 198), 3)
    pygame.draw.line(screen, GREEN, (350, 225), (340, 205), 3)
    pygame.draw.line(screen, GREEN, (355, 223), (345, 203), 3)

    # Листья слева
    pygame.draw.line(screen, GREEN, (380, 320), (200, 340), 4) # Верхняя ветвь
    pygame.draw.line(screen, GREEN, (200, 340), (180, 335), 3)
    pygame.draw.line(screen, GREEN, (205, 340), (185, 335), 3)
    pygame.draw.line(screen, GREEN, (230, 340), (210, 335), 3)
    pygame.draw.line(screen, GREEN, (235, 340), (215, 335), 3)

    pygame.draw.line(screen, GREEN, (380, 430), (250, 500), 4) # Нижняя ветвь
    pygame.draw.line(screen, GREEN, (250, 500), (230, 495), 3)
    pygame.draw.line(screen, GREEN, (255, 500), (235, 495), 3)
    pygame.draw.line(screen, GREEN, (280, 500), (260, 495), 3)
    pygame.draw.line(screen, GREEN, (285, 500), (265, 495), 3)

    #cоздаем доп экран и на нём рисуем панду
    panda_surface = pygame.Surface((350, 350), pygame.SRCALPHA)
    
    # Задняя часть 
    pygame.draw.ellipse(panda_surface, BLACK, [180, 130, 170, 160])
    # Передняя 
    pygame.draw.ellipse(panda_surface, BLACK, [70, 150, 130, 160])
    
    # Уши 
    pygame.draw.ellipse(panda_surface, BLACK, [205, 45, 55, 65])
    pygame.draw.ellipse(panda_surface, BLACK, [90, 55, 55, 65])

    # Белое пятно на спине
    pygame.draw.ellipse(panda_surface, WHITE, [190, 135, 130, 100])
    
    # Голова 
    pygame.draw.ellipse(panda_surface, WHITE, [90, 80, 160, 140])
    
    # Глаза 
    pygame.draw.circle(panda_surface, BLACK, (190, 140), 20)
    pygame.draw.circle(panda_surface, BLACK, (140, 140), 20)

    screen.blit(panda_surface, (mouse_x - 175, mouse_y - 175))

    pygame.display.flip()

pygame.quit()
```
